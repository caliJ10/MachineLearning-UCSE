# Consideraciones

"Perfil integrado": 
La superposición de algunos centenares de pulsos registrados, da el perfil integrado del pulso, el cual es normalmente estable y característico para cada púlsar.
El perfil integrado está constituido por una gran variedad de pulsos individuales, cada uno de los cuales tiene más de una componente: es la distribución estadística de estas componentes sobre un rango de longitudes, combinado con sus característicos anchos de banda y la distribución de probabilidad de sus intensidades lo que determina la forma repetitiva del perfil integrado.

Cada candidato consta de una selección de parámetros, que incluyen la medida de dispersión (DM) del púlsar, así como la relación señal-ruido (SNR) detectada en varios métodos de búsqueda.

#### Diagramas de caja

Diagrama de caja para las variables en el conjunto de datos.

```python
columns = [x for x in data.columns if x not in ["target_class"]]
length  = len(columns)
plt.figure(figsize=(13,20))
for i,j in itertools.zip_longest(columns,range(length)):
    plt.subplot(4,2,j+1)
    sns.boxplot(x="target_class", y=data[i],
            hue="target_class", palette=["m", "g"],
            data=data)
    plt.title(i)
    plt.subplots_adjust(hspace=.3)
    plt.axhline(data[i].mean(),linestyle = "dashed",color ="k",label ="Promedio para los datos")
    plt.legend(loc="best")
```

Como podemos observar en los distintos gráficos, en cada uno de ellos se presenta un gran número de valores atípicos, esto es algo a tener en cuenta a la hora de generar modelos para la predicción.

### Relación entre las variables

#### Cada variable se compara con todas las demás

```python
sns.pairplot(data=data,
             palette="husl",
             hue="target_class",
             vars=[
                "media del perfil integrado",
                "desviacion estandar del perfil integrado",
                "exceso de curtosis del perfil integrado",
                "sesgo del perfil integrado",
                "media de la curva DM-SNR",
                "desviacion estandar de la curva DM-SNR",
                "exceso de curtosis de la curva DM-SNR",
                "sesgo de la curva DM-SNR",
                "target_class"])
```

- Azul: target_class=1
- Rojo: target_class=0

#### Correlación entre las variables

```python
correlation = data.corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlation,annot=True,
            cmap=sns.color_palette("magma"),
            linewidth=2,edgecolor="k")
plt.title("CORRELACION ENTRE VARIABLES")
```

Observaciones de los datos obtenidos:

- El coheficiente de correlación entre la variable 'media del perfil integrado' y 'target_class' es de -0.67, por lo que podemos intuir que a medida que aumenta la variable 'media del perfil integrado', la variable de salida disminuye, es decir, tiende a ser 0.

- El coheficiente de correlación entre la variable 'exceso de curtosis de la curva DM-SNR' y 'target_class' es de -0.39, por lo que podemos intuir que a medida que aumenta la variable 'exceso de curtosis de la curva DM-SNRo', la variable de salida disminuye, es decir, tiende a ser 0.

- El coheficiente de correlación entre la variable 'exceso de curtosis del perfil integrado' y 'target_class' es de 0.79, por lo que podemos intuir que a medida que aumenta la variable 'exceso de curtosis del perfil integrado', la variable de salida aumenta, es decir, tiende a ser 1.

- El coheficiente de correlación entre la variable 'sesgo del perfil integrado' y 'target_class' es de 0.71, por lo que podemos intuir que a medida que aumenta la variable 'sesgo del perfil integrado', la variable de salida aumenta, es decir, tiende a ser 1.

- También podemos observar que algunas de las variables estan relaconadas estrechamente unas con otras, esto es algo a tener en cuenta si es que se requiere normalizar variables para la predicción.
